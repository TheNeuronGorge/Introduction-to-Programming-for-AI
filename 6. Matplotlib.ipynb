## Charts

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import matplotlib.image as mpimg
```

```python
# Sample data
x = [1, 2, 3, 4,13, 5]
y = [10, 15, 13, 1, 17, 20]
```

```python
# Create a simple line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Create a line plot with NumPy data
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with NumPy Data')

# Show the plot
plt.show()
```

```python
# Create a Pandas DataFrame
data = {'Year': [2015, 2016, 2017, 2018, 2019],
        'Revenue': [5000, 5500, 6000, 6500, 7000]}
```

```python
df = pd.DataFrame(data)

# Plotting Pandas DataFrame
df.plot(x='Year', y='Revenue', kind='line', marker='x')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Revenue (in $)')
plt.title('Revenue Over Years')

# Show the plot
plt.show()
```

### Kind
* 'line' (default): Line plot.
* 'bar': Vertical bar plot.
* 'barh': Horizontal bar plot.
* 'hist': Histogram.
* 'box': Box plot.
* 'kde': Kernel Density Estimation plot.
* 'density': Same as 'kde'.
* 'area': Area plot.
* 'scatter': Scatter plot.
* 'hexbin': Hexagonal bin plot.

### Marker
* 'o': Circle marker (default).
* 's': Square marker.
* '^': Triangle marker.
* '*': Star marker.
* '+': Plus marker.
* 'x': X marker.
* 'D': Diamond marker.
* '.': Point marker.

```python
data = np.random.normal(0,1, 1000)

# Create a histogram
plt.hist(data, bins=20, edgecolor='r')
```

### EdgeColor
* 'b': Blue
* 'g': Green
* 'r': Red
* 'c': Cyan
* 'm': Magenta
* 'y': Yellow
* 'w': White

## Histogram

```python
# Convert the image to grayscale (assuming it's a color image)
img_gray = np.mean(img2, axis=2)

# Flatten the 2D image array to a 1D array
img_flat = img_gray.flatten()

# Create a histogram
plt.hist(img_flat, bins=256, range=(0, 255), color='b')

# Add labels and title
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of the Image')

# Show the histogram
plt.show()
```

```python
# Load an image
image = cv2.imread('w3.png', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Plot the original and equalized images side by side
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```

```python

# Create a figure with subplots for the images and histograms
fig, axes = plt.subplots(1, 2, figsize=(15, 5))


# Plot the histogram of the original image
axes[0].hist(image.ravel(), bins=256, range=(0, 256), color='b')
axes[0].set_xlabel('Pixel Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of Original Image')

# Plot the histogram of the equalized image
axes[1].hist(equalized_image.ravel(), bins=256, range=(0, 256), color='b')
axes[1].set_xlabel('Pixel Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Equalized Image')

plt.tight_layout()
plt.show()
```

## Heat Map

```python
# Generate sample data
data = np.random.rand(10, 10)

# Create a heatmap
plt.imshow(data, cmap='viridis')
plt.colorbar()  # Add a colorbar for reference
plt.title('Basic Heatmap')
plt.show()
```

## 3D Charts

```python
# Generate sample data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='s')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot')
plt.show()
```
